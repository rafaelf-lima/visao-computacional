```python
import matplotlib.pyplot as plt

mao_cinza = cv2.imread("mao-escalada.jpg", cv2.IMREAD_GRAYSCALE)

# Plotar histograma dos pixels
plt.hist(mao_cinza.ravel(), bins=256, range=[0,256])
plt.title("Histograma de Intensidade")
plt.xlabel("Valor do Pixel")
plt.ylabel("Frequência")
plt.show()
```

```python
# Exibir a máscara
import matplotlib.pyplot as plt

# Aplicar threshold para segmentar a mão
_, mask = cv2.threshold(mao_cinza, 100, 255, cv2.THRESH_BINARY)

# Exibir a máscara no Colab
plt.figure(figsize=(6,6))
plt.imshow(mask, cmap="gray")
plt.axis("off")
plt.title("Máscara da Mão")
plt.show()

```
